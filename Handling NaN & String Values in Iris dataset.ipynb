```python
#Create NaN values in the dataset to understand how to handle NaN values
iris.iloc[2:4, 1:3] = np.nan
```

#we can either drop NaN values or we can fill NaN values

```python
#First we try with dropNa values
iris.dropna(inplace=True)
```

```python
iris.head()
```

```python
#Reset index for the dataset
iris.reset_index(drop=True, inplace=True)
iris.head()
```

```python
#Now create NaN again to apply the fillNaN method on the dataset
iris.iloc[2:4, 1:3] = np.nan
```

```python
iris.head()
```

```python
#Fill the NaN values with some value
#1st option is fill the NaN values with MEAN of the column
#2nd option is fill the NaN values with MOST OCCURING VALUE IN THE COLUMN
#Now try the 1st option
iris.SepalLengthCm.fillna(iris.SepalLengthCm.mean(), inplace=True)
iris.SepalWidthCm.fillna(iris.SepalWidthCm.mean(), inplace=True)
```

# Handling String Values

```python
#Create a gender column having string values
iris['Gender']='Female'
```

```python
iris.head()
```

```python
#Change some values from Female to Male in Gender Column
iris.iloc[1:10, 6]='Male'
```

```python
iris.head(15)
```

```python
#Create a function that change the string values to int accordingly (0 for Male and 1 for Female)
def f(s):
    if s == "Male":
        return 0
    else:
        return 1
    
iris['Sex'] = iris.Gender.apply(f)
del iris['Gender']
iris.head()
```
